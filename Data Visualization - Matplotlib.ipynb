```python
import matplotlib.pyplot as plt
company=['google','amzn','msf','fb']
turnover=[90,136,89,27]
profit=[40,46,3,12]
plt.bar(company,turnover,label='tunover')
plt.bar(company,profit,label='profit')
plt.legend()
plt.show()
```

```python
company=['google','amzn','msf','fb']
turnover=[90,136,89,27]
profit=[40,46,3,12]
plt.bar([0,1,2,3],turnover,width=0.4,label='revenue')
plt.bar([0.4,1.4,2.4,3.4],profit,width=0.4,label='profit')
plt.xlabel('company')
plt.xticks([0,1,2,3],company)
plt.legend()
plt.show()
```
